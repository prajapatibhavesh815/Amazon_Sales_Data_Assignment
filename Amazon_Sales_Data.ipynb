```python
# 1.what is the average rating for each product catogory?


avg_ratings = df.groupby('category')['rating'].mean()
print(avg_ratings)
```

```python
#2.What are the top rating_count products by category?

top_rated_products_by_catogory = df.groupby('category').apply(lambda x: x.nlargest(5,'rating_count'))
print(top_rated_products_by_catogory)
```

```python
#3.What is the distribution of discounted prices vs. actual prices?

#assuming df is your Dataframe containing product data
#and it has columns 'actual_price' and 'discounted_price

#plotting the distribution of actual prices
plt.figure(figsize=(10,6))
sns.histplot(df['actual_price'],bins=30,kde=True,color='blue',label='actual price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of actual prices')
plt.legend()
plt.show()

#Plotting the distribution of discounted prices
plt.figure(figsize=(10,6))
sns.histplot(df['discounted_price'],bins=30,kde=True,color='orange',label='discounted price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of discounted prices')
plt.legend()
plt.show()
```

```python
#4.How does the average discount percentage vary across categories?



#assuming df is your Dataframe containing product data
# and it has columns 'category', 'actual_price',and 'discounted_price'

#Calculate discount percentage for each product
df['discount_percentage'] = ((df['actual_price'] - df['discount_percentage']) / df['actual_price']) *100
avg_discount_percentage_by_category = df.groupby('category')['discount_percentage'].mean().reset_index()
print(avg_discount_percentage_by_category)
```

```python
#5. What are the most popular product names?


#Count the occurrences of each product name
product_name_counts = df['product_name'].value_counts()

#Get the top 10 most popular product names
top_popular_product_names = product_name_counts.head(10)

print(top_popular_product_names)
```

```python
#6. What are the most popular product keywords?

from collections import Counter
import re

#Function to extract keywords from product names
def extract_keywords(product_name):
    #Split product name into words
    words = re.findall(r'\b\w+\b',product_name.lower())
    #filter out common stopwords if necessary
    #words = [word for word in words if word not in stopwords]
    return words

#Apply the function to each product name and combine the results
keywords_list = df['product_name'].apply(extract_keywords).sum()

#Count the occurences of each keyword

keyword_counts = Counter(keywords_list)

#get the top 10 most popular keywords_list
top_popular_keywords = keyword_counts.most_common(10)

print(top_popular_keywords)
```

```python
#7. What are the most popular product reviews?

#Assuming df is your Dataframe containing product review data
# and it has columns named 'product_name' and 'review_text'


#count the occurences of each review text
review_counts = df['review_title'].value_counts()

#Get the top 10 most popular reviews
top_popular_reviews = review_counts.head(10)
print(top_popular_reviews)
```

```python
#8. What is the correlation between discounted_price and rating?


#assuming df is your dataframe containing product data
# and it has columns 'discounted_price' and 'rating'

#Calculate the correlation coefficient between 'discount_price' and 'rating'

correlation = df['discounted_price'].corr(df['rating'])

print("correlation coefficient between discounted_price and rating:",correlation)

#plotting the data
plt.figure(figsize=(8,6))
plt.scatter(df['discounted_price'],df['rating'],alpha=0.5)
plt.title('correlation between discounted price and rating')
plt.xlabel('discounted price')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
#9. What are the Top 5 categories based on the highest ratings?

#Calculate the average rating for each category
avg_rating_by_category = df.groupby('category')['rating'].mean()

# Sort the categories based on average rating in descending order and get the top 5
top_categories = avg_rating_by_category.sort_values(ascending=False).head(5)

print("Top 5 categories based on the highest ratings:")
print(top_categories)
```
